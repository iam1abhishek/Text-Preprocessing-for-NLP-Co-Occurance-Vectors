```python
import numpy as np
from collections import defaultdict
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text="Co-occurance vectors capture word relationships in a text corpus. THEY ARE USED for various NLP tasks."
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences= sent_tokenize(text)
tokenized_sentences=[word_tokenize(sentence.lower()) for sentence in sentences]
print(tokenized_sentences)
print()
```

# build co-occurance matrix

```python
window_size=2
co_occurance_matrix=defaultdict(lambda: defaultdict(int))

for sentence in tokenized_sentences:
  for i, target_word in enumerate(sentence):
    for j in range(max(0,i-window_size),min(len(sentence),i+window_size+1)):
      if i !=j:
        context_word=sentence[j]
        co_occurance_matrix[target_word][context_word]+=1

```

# Print co-occurance matrix

```python
for target_word, context_words in co_occurance_matrix.items():
  print(target_word,context_words)
print()
```
